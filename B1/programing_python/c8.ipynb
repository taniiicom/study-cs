```python
"""
    課題1: 辞書の読み込みと統計値の計算
"""
import mod8th

STUDENT_NUM = 22140026

mod8th.init(STUDENT_NUM)

scores = mod8th.getPractice1()

sum = 0
max_key = None
min_key = None
for k, v in scores.items():
    if (max_key == None) or (min_key == None):
        max_key = min_key = k
    else:
        sum += v
        if v < scores[min_key]:
            min_key = k
        if v > scores[max_key]:
            max_key = k

ave = sum / len(scores)
print("最大値: ", max_key, "->", scores[max_key])
print("最小値: ", min_key, "->", scores[min_key])

sum_d = 0
for v in scores.values():
    sum_d += (v - ave)**2

print("平均値: ", ave)
print("標準偏差: ", (sum_d/len(scores))**(1/2))
```

```python
"""
    課題2
"""
import mod8th

STUDENT_NUM = 22222222

mod8th.init(STUDENT_NUM)

p2d = mod8th.getPractice2Dict()

text = mod8th.getPractice2()

splited = list(text)

for i in range(0, len(splited)):
    try:
        splited[i] = p2d[splited[i]]
    except:
        pass

print("log: ", p2d)
print("log: ", text)
print("log: ", splited)

text2 = ""
for t in splited:
    text2 += t

print(text2)

splited2 = text2.split("-")

splited3 = []
for t in splited2:
    splited3.append(int(t))

print("平均値: ", sum(splited3)/len(splited3))
print("最大値: ", max(splited3))
print("最小値: ", min(splited3))
```

```python
"""
    課題3: 「円周率の近似」
"""
def myPi(n):
    En = 0
    for i in range(0, n+1):
        ei = (16**(-i))*(4/(8*i+1)-(2/(8*i+4))-(1/(8*i+5))-(1/(8*i+6)))
        En += ei
    return En

print(myPi(int(input("n="))))
```

```python
"""
    課題4: tupleとlistの違い
"""
# tupleとは: 
# tupleは, listと似ているが, listとは異なり, 
# 同じオブジェクトidのまま, 要素をあとから変更できない.  


# 課題の説明: 
# list と tuple にそれぞれ, 追加/挿入/削除したときにエラーが発生するかを確認する.  
list_camp = ["南大沢", "日野", "荒川"]
tuple_camp = ("南大沢", "日野", "荒川")

try:
    list_camp.append("目黒")
    list_camp.insert(0, "目黒")
    list_camp.remove("日野")
except:
    print("【list】への追加/挿入/削除でエラーが発生...")
else:
    print("【list】への追加/挿入/削除に成功")

try:
    tuple_camp.append("目黒")
    tuple_camp.insert(0, "目黒")
    tuple_camp.remove("日野")
except:
    print("【tuple】への追加/挿入/削除でエラーが発生...")
else:
    print("【list】への追加/挿入/削除に成功")


# 考察: tupleでは, 同じオブジェクトidのまま, 要素を変更することができない

# 使い道: 
# 緯度経度や47都道府県や"100以下の素数"のように後から変更することがないデータや, dictのkeyに利用したい場合にtupleを使用する.  

```
